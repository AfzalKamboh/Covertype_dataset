**Distributing the Dataset for training and testing That used during train_test_split function.**

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

**Perceptron Accuracy**

```python
from sklearn.linear_model import Perceptron
p=Perceptron(random_state=42)
p.fit(X_train,y_train)
```

**Training and Testing Data accuracy score in Perceptron.**

```python
from sklearn.metrics import accuracy_score

predictions_train = p.predict(X_train)
predictions_test = p.predict(X_test)
train_score = accuracy_score(predictions_train, y_train)
print("score on train data: ", train_score)
test_score = accuracy_score(predictions_test, y_test)
print("score on test data: ", test_score)
```

**Final Accuracy Score of Perceptron.**

```python
p.score(X_train, y_train)
```

**Data Spliting for KNeighbours Algorithm.**

```python
X=data.loc[:,'Elevation':'Soil_Type40']
y=data['class']
```

```python
rem=['Hillshade_3pm','Soil_Type7','Soil_Type8','Soil_Type14','Soil_Type15',
     'Soil_Type21','Soil_Type25','Soil_Type28','Soil_Type36','Soil_Type37']
```

**Removing the Columns**

```python
X.drop(rem, axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

**Declaring Packages that are important.**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import numpy as np
```

**Training of KNN**

```python
neighbors = np.arange(1,7)
train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)
    
    #Fit the model
    knn.fit(X_train, y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)
    
    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(X_test, y_test)
```

**KNN algorithm Test and Training Accuracy Visualization.**

```python

plt.figure(figsize=(10,6))
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

**Fitting the training and Testing data for KNN**

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
           metric_params=None, n_jobs=1, n_neighbors=5, p=2,
           weights='uniform')
```

**Final Accuracy of KNN Classifier Algorithm.**

```python
from sklearn.metrics import accuracy_score
Accuracy=knn.score(X_test,y_test)
print('KNN Accuracy:',Accuracy)
```

**Calculating the Accuracy of All other Algorithms.**

```python
from scipy.stats import zscore
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
MLA = []
Z = [DecisionTreeClassifier() , LogisticRegression(),Perceptron(),SVC(),KNeighborsClassifier()]
X = ["DecisionTreeClassifier" , "LogisticRegression","Perecptron","SVM","KNeighborsClassifier"]

for i in range(0,len(Z)):
    model = Z[i]
    model.fit( X_train , y_train )
    pred = model.predict(X_test)
    MLA.append(accuracy_score(pred , y_test))
```
